```python
# now i will split the feautres and target
X_data=data.drop(columns=['Chronic Kidney Disease: yes'])
```

```python
Y_data=data.iloc[:,-1:]
```

```python
# now i will split the data into train and test
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y_data,test_size=0.3,random_state=2)
```

# Now i use the different classification algorithm for better accuracy

1.Support vector machine classifier

```python
from sklearn.svm import SVC
```

```python
model1=SVC()
```

```python
model1.fit(X_train,Y_train)
```

```python
prediction1=model1.predict(X_test)
```

```python
# Now i check the accuracy using different accuracy metrics
```

```python
# 1.accuracy_score
from sklearn.metrics import accuracy_score
```

```python
ans1=accuracy_score(Y_test,prediction1)
```

```python
print(ans1)
```

```python
# 2.balanced_accuracy_score
from sklearn.metrics import balanced_accuracy_score
```

```python
ans2=balanced_accuracy_score(Y_test,prediction1)
```

```python
print(ans2)
```

```python
# now i apply the cross-validation
from sklearn.model_selection import cross_val_score
```

```python
result=cross_val_score(model1,X_train,Y_train,cv=8,scoring='accuracy')
```

```python
print(result)
```

```python
result.mean()
```
